```
from plotly.offline import plot
import plotly.graph_objs as go
import pandas as pd
import numpy as np
from datetime import datetime
```

```
df  = pd.read_csv("/content/sample_data/alipur.csv",error_bad_lines=False)
```

```
df
```

```
df['From Date'].replace(to_replace=['None',np.nan,'Prescribed Standards','Exceeding Standards','Remarks','From Date'], value='01-01-9999 00:00', inplace=True)

for i in range(0,len(df['From Date'])):
 print(df['From Date'][i]) 
 df['From Date'][i] = datetime.strptime(df['From Date'][i],"%d-%m-%Y %M:%S")

```

```
df['To Date'].replace(to_replace=['None',np.nan,'Prescribed Standards','Exceeding Standards','Remarks','From Date','To Date',' '], value='01-01-9999 00:00', inplace=True)

for i in range(0,len(df['To Date'])):
 print(df['To Date'][i]) 
 df['To Date'][i] = datetime.strptime(df['To Date'][i],"%d-%m-%Y %M:%S")
```

```
df['SR'].replace(to_replace=['None',np.nan,'RH','0-200','NOx','0-180','Ozone','Eth-Benzene'], value=0, inplace=True)
df['SR']=df['SR'].astype(float)
```

```
df['BP'].replace(to_replace=['None',np.nan,'NO','0-4','CO','MP-Xylene','0-200'], value=0, inplace=True)
df['BP']=df['BP'].astype(float)
```

```
df['AT'].replace(to_replace=['None',np.nan,'RF','0-80','S02','Xylene','SO2'], value=0, inplace=True)
df['AT']=df['AT'].astype(float)
```

```
df['PM2.5'].replace(to_replace=['None',np.nan,'WD','0-200','NH3','Toluene'], value=0, inplace=True)
df['PM2.5']=df['PM2.5'].astype(float)
```

```
df['PM10'].replace(to_replace=['None',np.nan,'RH','0-200','NO2','Benzene'], value=0, inplace=True)
df['PM10']=df['PM10'].astype(float)
```

```
pm10_series = df.loc[:,['To Date','PM10']]
print(pm10_series.describe())
pm10_series.head()
```

```
pm10_series = pm10_series[pm10_series['PM10']!=0] 
pm10_series.describe()
```

```
pm10_series = pm10_series[pm10_series['To Date']!='01-01-9999 00:00'] 
pm10_series.describe()
```

```
time = pm10_series['To Date']

y = pm10_series['PM10']

plot(go.Figure(data = [go.Scatter(x = time, y = y)]))
```
