```
from selenium import webdriver
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
```

```
wd = webdriver.Chrome('/usr/lib/chromium-browser/chromedriver',options=options)
```

```
columns = ['Date','Time','AM/PM','Temperature(Fh)','Dew_Point(Fh)','Humidity(%)','Wind','Wind Speed(mph)','Wind Gust(mph)','Prssure(in)','Preception(in)','Condition']
data = pd.DataFrame(columns=columns)
data
```

```
url = "https://www.wunderground.com/history/daily/in/new-delhi/VIDP/date/"

start_date = date(2017, 1, 1) 
end_date = date(2019, 12, 31)   

delta = end_date - start_date    
row=0
for i in range(delta.days + 1):
    
    day = start_date + timedelta(days=i)
    
    print(url+str(day))
        #mat-table ng-star-inserted
    wd.get(url)
    
    delay = 10# seconds

    temp=[]

    Date=[]
    Time=[]
    AMPM=[]
    Temperature=[]
    Dew_Point=[]
    Humidity=[]
    Wind=[]
    Wind_Speed=[]
    Wind_Gust=[]
    Prssure=[]
    Preception=[]
    Condition=[]

    try:
        table = WebDriverWait(wd, delay).until(ec.presence_of_element_located((By.XPATH, '//*[@id="inner-content"]/div[2]/div[1]/div[5]/div[1]/div/lib-city-history-observation/div/div[2]/table')))
        print ("Fecthed Date for "+str(day))
        
        dlist = (table.text).splitlines()

        for  i in range(10,len(dlist)):
          dstr = dlist[i]
          dlst = dstr.split(' ')
          
          Date.append(day)
          Time.append(dlst[0])
          AMPM.append(dlst[1])
          Temperature.append(float(dlst[2]))
          Dew_Point.append(float(dlst[4]))
          Humidity.append(float(dlst[6]))
          Wind.append(dlst[8])
          Wind_Speed.append(float(dlst[9]))
          Wind_Gust.append(float(dlst[11]))
          Prssure.append(float(dlst[13]))
          Preception.append(float(dlst[15]))
          Condition.append(dlst[17])
          
          '''obj = {
                'Date':day ,
                'Time':dlst[0],
                'AM/PM':dlst[1],
                'Temperature(Fh)':dlst[2],
                'Dew_Point(Fh)':dlst[4],
                'Humidity(%)':dlst[6],
                'Wind':dlst[8],	
                'Wind Speed(mph)':dlst[9],
                'Wind Gust(mph)':dlst[11],
                'Prssure(in)':dlst[13],
                'Preception(in)':dlst[15],
                'Condition':dlst[17] 
                }'''
         
  
        obj = {
            'Date':Date,
            'Time':Time,
            'AM/PM':AMPM,
            'Temperature(Fh)':Temperature,
            'Dew_Point(Fh)':Dew_Point,
            'Humidity(%)':Humidity,
            'Wind':Wind,	
            'Wind Speed(mph)':Wind_Speed,
            'Wind Gust(mph)':Wind_Gust,
            'Prssure(in)':Prssure,
            'Preception(in)':Preception,
            'Condition':Condition 
            }

        df=pd.DataFrame(obj)          
        temp=[data,df]
        data=pd.concat(temp)
        print ("Saved Date for "+str(day))
           
    except Exception as e :
        print (str(e))  

print(data)

```
